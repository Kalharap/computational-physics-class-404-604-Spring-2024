```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, n):
    result = np.pi/2
    for i in range(1, n+1):
        result += 2*(np.sin(i*x)/i)
    return result

x_val = np.arange(-3.14,3.14,0.01)

plt.figure(figsize=(10, 6))

for n in [11, 21, 31]:
    y_val = np.vectorize(lambda x: f(x, n))(x_val)
    plt.plot(x_val, y_val, label=f'n = {n}')

plt.title('Plot of f(x) for Different Values of n')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
# Recursion method fib2, 
def fib2(n):      # extremely simple, but exponentially running time, very slow for large n, n cannot too large (stack limit)
    if n == 0: return 0
    elif n == 1: return 1
    else: return fib2(n-1)+fib2(n-2)
```
