```python
import matplotlib.pyplot as plt
import numpy as np

def f1(x):
    return (10. - x*x) / x

def f2(x):
    return (-1 + np.sqrt(1. + 4. * 3. * x * 57)) / 6. / x

x_values = np.linspace(0.1, 10, 100)  

y1_values = f1(x_values)
y2_values = f2(x_values)

plt.plot(x_values, y1_values, label='f1(x)')
plt.plot(x_values, y2_values, label='f2(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphs of f1(x) and f2(x)')
plt.legend()

plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


def f1(x1,x2):
    return (x1*x1+x1*x2-10.)

def f2(x1,x2):
    return (x2+3*x1*x2*x2-57)

def f1x1(x1,x2,dx):
    return (f1(x1 + dx/2, x2) - f1(x1 - dx/2, x2)) / dx

def f1x2(x1, x2, dx):
    return (f1(x1, x2 + dx/2) - f1(x1, x2 - dx/2)) / dx
    
def f2x1(x1,x2,dx):
    return (f2(x1 + dx/2, x2) - f2(x1 - dx/2, x2)) / dx
    
def f2x2(x1,x2,dx):
    return (f2(x1, x2 + dx/2) - f2(x1, x2 - dx/2)) / dx
    
def dx1(f1,f2,f1x1,f1x2,f2x1,f2x2,x1,x2,dx):
    return ((f1x2(x1,x2,dx)*f2(x1,x2))-(f2x2(x1,x2,dx)*f1(x1,x2)))/((f1x1(x1,x2,dx)*f2x2(x1,x2,dx))-(f1x2(x1,x2,dx)*f2x1(x1,x2,dx)))

def dx2(f1,f2,f1x1,f1x2,f2x1,f2x2,x1,x2,dx):
    return ((f2x1(x1,x2,dx)*f1(x1,x2))-(f1x1(x1,x2,dx)*f2(x1,x2)))/((f1x1(x1,x2,dx)*f2x2(x1,x2,dx))-(f1x2(x1,x2,dx)*f2x1(x1,x2,dx)))
    
x1 = 1.0
x2 = 1.0
eps = 1e-12
max_iterations = 20
dx = 1e-6 

for i in range(max_iterations):
    f1_val = f1(x1, x2)
    f2_val = f2(x1, x2)
    if abs(f1_val) < eps and abs(f2_val) < eps:
        break
    x1_new = x1 + dx1(f1,f2,f1x1,f1x2,f2x1,f2x2,x1,x2,dx)
    x2_new = x2 + dx2(f1,f2,f1x1,f1x2,f2x1,f2x2,x1,x2,dx)
    x1 = x1_new
    x2 = x2_new

print("Roots of (x1, x2):", x1, x2)
```
