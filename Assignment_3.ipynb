## Question 01

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def derifw(y,x,dx):
    return (y(x+dx)-y(x))/dx
```

```python
x=(np.pi)/4
dx=0.1
print('forward difference',derifw(np.cos,x,dx))
print('analytical result',-np.sin(x))
```

```python
x=(np.pi)/4
dxarr=[]
yarr=[]
for dx in 10**np.arange(-16, -0.5, 0.1):
    y=derifw(np.cos,x,dx)
    dxarr.append(dx)
    yarr.append(y)
    
plt.plot(dxarr, abs(yarr+np.sin(x)), 'b-', lw=2)    
plt.xscale('log')
plt.yscale('log')
plt.xlabel('dx')
plt.ylabel('Error')

plt.show()
```

The optimal value for n =8, as we have two types of erros, truncation error and the machine precission, we have to take the optimum value for n, that dicreases the error as possible

## Question 2

```python
def dericd(y,x,dx):
    return (y(x+dx/2)-y(x-dx/2))/dx
```

```python
x=(np.pi)/4
dxarr1=[]
yarr1=[]
for dx in 10**np.arange(-16, -0.5, 0.1):
    y1=dericd(np.cos,x,dx)
    dxarr1.append(dx)
    yarr1.append(y1)
    
plt.plot(dxarr1, abs(yarr1+np.sin(x)), 'b-', lw=2)    
plt.xscale('log')
plt.yscale('log')
plt.xlabel('dx')
plt.ylabel('Error')

plt.show()
```

The Round off error is almost the same for both methods, but the difference is in thr trucation error, in the forward scheme, the leading term for truncation error is in first order, but in the central sceme it in the second order, therefore the error drastically reduces with the central scheme.

The truncation error for forward difference is  $y''\times h/2$, for central difference is  $y'''\times h^2/24$
