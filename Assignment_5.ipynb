```python
import numpy as np
import matplotlib.pyplot as plt

g = 9.80
tend = (2*y[1])/g
h = 0.001
y = [0 ,10.]
t = 0

def f(t, y):
    f = np.zeros(2)
    f[0] = y[1]  # Velocity
    f[1] = -g      # Acceleration due to gravity
    return f

def eu(t, h, y):
    g = f(t, y)
    for i in range(2):
        y[i] = y[i] + (h * g[i])  # Euler method
    return y

xp = []
yp = []

while t < tend:
    if (t + h) > tend:
        h = tend - t
    y = eu(t, h, y)
    xp = np.append(xp, y[0])
    t = t + h


plt.plot(np.linspace(0, tend, len(xp)), xp)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('ball Motion with Euler Method')
plt.show()
```
